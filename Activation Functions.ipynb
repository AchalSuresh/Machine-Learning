```python
import numpy as np
import matplotlib.pyplot as plt


#Sigmoid function
def sigmoid(x):
  return 1.0 / (1.0 + np.exp(-x))

#Deriavtive of function
def sigmoid_derivative(x):
    s = sigmoid(x)
    ds = s*(1-s)
    return ds


#Providing the range of input and calling function
xs = np.arange(-10,10,0.1)
ys1= sigmoid(xs)
ys2 = sigmoid_derivative(xs)

#PLotting the function
ax = plt.gca()
ax.yaxis.grid(True)
plt.axvline(0.0, color='k')
plt.plot(xs,ys1,label='Sigmoid Function')
plt.plot(xs,ys2,label='Derivative')
plt.legend()
plt.show()
```

Tangent Action Function and its derivative

```python
#Tangent function
import numpy as np 
import matplotlib.pyplot as plt

#Hyperbloic Tangent function
in_array = np.arange(-10,10,0.1)
out_array = np.tanh(in_array)

#Derivation of the function
def tanh_deriv(x):
    return 1.0 - np.tanh(x)**2

#Triggering the derivative
def_array = tanh_deriv(in_array)

#Plotting the function
ax = plt.gca()
ax.yaxis.grid(True)
plt.axvline(0.0, color='k')
plt.plot(in_array,out_array, label= "Hyperbolic tangent function")
plt.plot(in_array,def_array,label = "Derivative")
plt.legend()
plt.show()
```

Relu Activation Function

```python
#Relu function
def Relu(x):
  if x  >   0 or x ==0:
    return x
  else:
    return 0

#Derivative of function
def diff_Relu(x):
  if x  >   0 or x ==0:
    return 1
  else:
    return 0

#Input and calling the functions
xs= np.arange(-10,10,0.1)
ys1 = []
ys2 = []
for x in xs:
    ys1.append(Relu(x))
    ys2.append(diff_Relu(x))
    
#Plotting the function    
ax = plt.gca()
ax.yaxis.grid(True)
plt.axvline(0.0, color='k')
plt.plot(xs,ys1,label='Relu function')
plt.plot(xs,ys2,label='Derivative of function')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import misc
#Relu function
def Relu(x):
  if x  >   0 or x ==0:
    return x
  else:
    return 0

#Derivative of function
def diff_Relu(x):
  if x  >   0 or x ==0:
    return 1
  else:
    return 0
x = np.arange(-10, 10, 0.001)
ys1 =[]
ys2 = []
ys3 =[]
for x in xs:
    ys1.append(Relu(x))
    ys2.append(diff_Relu(x))
    ys3.append(misc.derivative(Relu, x, dx=0.001))
ax = plt.gca()
ax.yaxis.grid(True)
plt.axvline(0.0, color='k')
plt.plot( xs,ys2, markerfacecolor='blue', color='skyblue', linewidth=1,label='y1')
plt.plot( xs,ys3, marker='', color='olive', linewidth=2,label='y2')
plt.legend()
```

```python
import pandas as pd
#loading the dataset
covid = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv",sep=",")
#groupinhg the datset by country
covid= covid.groupby('Country/Region').sum()
covid.reset_index()

#sorting the dataset and selecting the 10 countires with the  highest cases
covid_sorted= covid.sort_values(by='3/28/20', ascending=False)
covid_10=covid_sorted.iloc[0:10,2:]
covid.head()
```
